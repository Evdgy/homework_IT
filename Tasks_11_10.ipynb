```python
# Find all words in a string that start with a capital letter.
import re

# define the input string
input_str = "Hello world, this is a Regex example with Python."

# define the regex pattern to match capital letters
pattern = r'\b[A-Z]\w*\b'

# use re.findall() to find all matches of the pattern in the input string
matches = re.findall(pattern, input_str)

# print the matches
print(matches)
```

```python
# Replace all instances of "is" with "was" in the string.
sentence = "This is a test. This is only a test."
pattern_1 = r'\bis\b'
replacement = 'was'
sentence_2 = re.sub(pattern_1, replacement, sentence)
print(sentence_2)
```

```python
# Extract All URLs from a Text
sentence_2 = """
Visit https://www.example.com or follow http://example.org.
You can also check out https://another-example.com/path?arg=val for more details.
"""
links = re.findall(r'http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*/(\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+', sentence_2) # r -  to detect that there is no need in / 
print(links)


```

```python
# Find all hashtags in a comment
comment = "Loving the #sunset and #nature! #beautiful_day"
pattern_3 = r"#\w+"
new_comments = re.findall(pattern_3, comment)
print(new_comments)
```

```python
#Remove the repeated words
sentence_5 = "This is is a test test string."
pattern_5 = r'\b(\w+)(?:\W+\1\b)+'
new_sentence = re.sub(pattern_5, r'\1', sentence_5)
print(new_sentence)
```
