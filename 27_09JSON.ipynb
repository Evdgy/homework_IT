```python
#1. Write a Python program to convert JSON data to Python object.
import json
data = '{"country" : "Argentina", "id" : 2, "location" : "South America"}'
x = json.loads(data)
print("\nJSON information:")
print(x)
print("\ncountry: ",x["country"])
print("id: ",x["id"])
print("location: ",x["location"])
```

```python
# Write a Python program to convert Python dictionary object (sort by key) to JSON data. Print the object members with indent level 4.
import json
py_dict_1 =  {"country": "Japan", "id": 3, "short":"JP"} 
py_dict_2 = {"country": "Argentina","id": 2, "short":"AR",}
print("Dictionary:", py_dict_1, py_dict_2)
print("\nJSON data:")
print(json.dumps(py_dict_1, sort_keys=True, indent=4))
print(json.dumps(py_dict_2, sort_keys=True, indent=4))
```
