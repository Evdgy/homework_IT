```python
# Write a Python program to read a given CSV files with initial spaces after a delimiter and remove those initial spaces.
import csv 
print('\nFile with initial spaces:\n')
with open('employees.csv', 'r') as csvfile:
    reader = csv.reader(csvfile, skipinitialspace=False)
    for row in reader:
        print(', '.join(row))
print('\nFile without initial spaces\n')
with open('employees.csv', 'r') as csvfile:
    reader = csv.reader(csvfile, skipinitialspace=True)
    for row in reader:
        print(', '.join(row))
```

```python
# Write a Python program that reads a CSV file and remove initial spaces, quotes around each entry and the delimiter.
import csv     
class my_dialect(csv.Dialect):
    lineterminator = '\n'
    delimiter = '|'
    skipinitialspace=True
    quotechar = '"'
    quoting=csv.QUOTE_ALL
with open('countries.csv', 'r') as f:
    reader = csv.reader(f, dialect=my_dialect)
    for row in reader:
        print(row[0].split())
```

```python
# Write a Python program to create an object for writing and iterate over the rows to print the values

import csv
import sys
with open('temp.csv', 'wt') as file:
    writer = csv.writer(file)
    writer.writerow(('ID_1', 'ID_2', 'date'))
    for num in range(3):
        row = (
            num + 1,
            chr(ord('a') + num),
            '01/{:02d}/2019'.format(num + 1),)
        writer.writerow(row)
print(open('temp.csv', 'rt').read())
```

```python
# Write a Python program to write a Python list of lists to a csv file. After writing the CSV file read the CSV file and display the content.
import csv

listOfLists = [[1, ' ', "Mary", ' ' "Teacher"],[2, ' ' "Sam", ' ' 'Economist'],[3, ' ' 'Chris',' ' 'Programmer']]
  

myFile = open('employees.csv', 'w')
writer = csv.writer(myFile)
writer.writerow(['Number', ' ' 'Name' ' ','Profession'])

for data_list in listOfLists:
   writer.writerow(data_list)

myFile.close()
myFile = open('employees.csv', 'r')
print("The content of the csv file is:")
print(myFile.read())
myFile.close()
```

```python
# Write a Python program to write a Python dictionary to a csv file. After writing the CSV file read the CSV file and display the content.
import csv
csv_columns = ['NUMBER','1', '2', '3']
dict_data = {'NUMBER':['exact_num'],
    '1':[20],
    '2':[21],
    '3':[22],}
csv_file = "departments.csv"
try:
   with open(csv_file, 'w') as csvfile:
       writer = csv.DictWriter(csvfile, fieldnames=csv_columns)
       writer.writeheader()
       for data in dict_data:
           writer.writerow(dict_data)
except IOError:
   print("I/O error")
data = csv.DictReader(open(csv_file))
print("CSV file as a dictionary:\n")
for row in data:
   print(row)
```
